**1. Creating and Plotting Graphs**

Matplotlib simplifies the process of creating visualizations. Hereâ€™s a quick and easy example of how to create and plot a basic line graph:

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating the plot
plt.plot(x, y)

# Displaying the plot
plt.title('Simple Line Graph')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
````

**2. Different Chart Types**

Matplotlib supports a variety of chart types. Here are examples of three common types:

*   **Bar Chart**: Useful for comparing quantities corresponding to different groups.

````python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 8]

# Creating the bar chart
plt.bar(categories, values, color='blue')

# Displaying the chart
plt.title('Simple Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

*  **Scatter Plot:** Ideal for showing the relationship between two variables and detecting any trends.

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 2, 4, 5]

# Creating the scatter plot
plt.scatter(x, y, color='red')

# Displaying the plot
plt.title('Simple Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
````

*  **Histogram**: Used for showing frequency distributions.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
# Generates 1000 random values from a normal distribution centered at 0 with a standard deviation of 1, spreading mostly between -3 and 3.
data = np.random.normal(0, 1, 1000)
# data will fall between -3 and 3 because. 68% of the data falls within one standard deviation of the mean. In your case, since the standard deviation is 1, this would mean between -1 and 1.
# 95% of the data falls within two standard deviations of the mean. So, between -2 and 2.
# 99.7% of the data falls within three standard deviations of the mean, which would be between -3 and 3.

# Creating the histogram
plt.hist(data, bins=30)

# Displaying the histogram
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

**3. Modification of Charts for Better Understanding and Presentation**

To make charts more understandable and visually appealing, you can modify them with titles, labels, legends, and color schemes. Hereâ€™s how to enhance a basic chart:

````python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100) # generates 100 equally spaced values between 0 and 10
y = np.sin(x) # This calculates the sine of each value in the array x

# Creating the plot with modifications
plt.plot(x, y, label='Sine Wave', color='purple')

# Enhancing the plot
plt.title('Enhanced Line Graph with Labels and Legend')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

# Customizing with grid
plt.grid(True) # This line of code enables a grid in the plot.

# Displaying the plot
plt.show()
````

This modified plot includes a title, axis labels, a legend, and a grid, making it easier to understand and visually appealing.

**Key Principles of Effective Data Visualization**

Creating impactful data visualizations involves more than just presenting numbers and stats. Here are essential points to consider to enhance comprehension and engagement:

*   Purpose-Driven Design: Tailor your visualization to augment the understanding of the data. Start by defining the purpose of the visualization and the story you want to tell.

*   Strategic Placement: Since viewers often look at the top left first, place the most important data where it will be seen immediately. This ensures key insights are not overlooked.

*   Appropriate Chart Type: Select chart types that best represent the nature and relationships within the data. Misrepresentation can lead to incorrect interpretations.

*   Color and Contrast: Use color wisely to draw attention, signify categories, or denote quantities. High contrast can emphasize important points, while a harmonious color scheme can make the visualization pleasing to the eye and easier to decode.

*   Narrative Support: Whenever possible, accompany your visualizations with a voice or textual narrative. This helps in guiding the viewer through the data, providing context, and emphasizing key insights.

*   Simplicity Over Complexity: Avoid clutter and focus on clarity. Simplify the visualization by removing unnecessary elements. A clean and clear presentation enhances understanding.

*   Interactive Elements: If applicable, include interactive elements such as tooltips, clickable legends, or sliders to allow viewers to explore the data in more depth. This can make complex data more accessible.

*   Accessibility: Ensure your visualizations are accessible to all audiences, including those with visual impairments. Use alt text for charts, and consider color blindness when choosing color palettes.

# Facts and Dimensions

Matplotlib is a powerful tool that enhances the representation of raw data through various graphical formats. It allows users to transform data stored in lists, arrays, and other formats into meaningful visual narratives. Here's a quick overview as we embark on learning about Matplotlib:

Types of Data: Understanding the basic elements of chart creation is crucial. There are two primary types of data:
*  Facts: These are numerical data points that provide quantifiable measures, such as the height of students. Other examples: Withdrawal amount, Account balance after withdrawal, Transaction charge amount
*  Dimensions: These are metadata that give context to facts, like the gender of students, allowing for a more nuanced analysis. Other examples: Customer ID, ATM ID, Date of withdrawal


Both facts and dimensions play vital roles in crafting insightful visualizations. For instance, by using both the factual data (height) and dimensional data (gender), one can analyze and visualize the distribution of student heights by gender, offering deeper insights into the data set. This foundational knowledge is essential for effectively starting data analysis with Matplotlib.

##### Bar Chart - Plot sales across each product category

1. Adding labels to Axes
2. Reducing the bar width
3. Giving Title to the chart
4. Modifying the ticks to show information in (million dollars)

````python
import numpy as np
import matplotlib.pyplot as plt

product_category = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52] )

# plotting bar chart and setting bar width to 0.5 and aligning it to center
plt.bar(product_category, sales, width= 0.5, align='center', edgecolor='Orange',color='cyan')

# Adding and formatting title # fontweight means thickness, 700 means bold
plt.title("Sales Across Product Categories\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Product Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Sales", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})

# Modifying the ticks to show information in (million dollars)
ticks = np.arange(0, 6000000, 1000000)
labels = ["{}M".format(i//1000000) for i in ticks]
plt.yticks(ticks, labels)

plt.show()
````

Outliers:

Additional Definitions:
*  Interquartile Range (IQR): The range between the 25th percentile (Q1) and the 75th percentile (Q3). It measures the middle 50% spread of the data.

# **Choosing Plot Types**

Each of the plot types is good at communicating a specific type of information. Which means, in certain situations, certain plot types are preferred over the others. So, how do you select the best possible plot type in a given situation? To answer this question, you need to first define the objective of creating a plot. A good visualisation, along with the right type of graph, presents the relationship between different variables effectively and allows you to analyse them at a quick glance.

**Comparison**

These charts can be used when you want to compare one set of values with other sets of values. The objective is to differentiate one particular set of values from the other sets, for example, quarterly sales of competing phones in the market. The following two types of charts are used to show a comparison:  

* Column chart
* Bar chart

**Composition**

You would need to use a composition chart to display how the various elements make up the complete data. Composition charts can be static, which shows the composition at a particular instance of time, or dynamic, which shows the changes in the composition over a period of time. Two of the popular composition charts are as follows:
* Pie/ Doughnut Chart
* Stacked Column chart

The pie chart is by far the most common way to represent static composition, while the stacked column chart can be used to show the variation of composition over a period of time.

**Relationship**

A relationship chart helps in visualising the correlation between variables. It can help in answering questions such as â€˜Is there a correlation between the amount spent on marketing and the sales revenue?â€™ and â€˜How does the gross profit vary with the change in offers?â€™. Two of the most common types of charts used to visualise relationships between variables are as follows:
* Scatter plot
* Bubble plot

A scatter plot can help correlate two variables, whereas a bubble chart adds one more dimension, i.e., the size of the bubble (usually indicative of the frequency of occurrence of that particular data point)

**Distribution**

A distribution chart tries to answer the question â€˜How is the data distributed?â€™. For example, suppose you asked everyone their age in a survey. Using a distribution chart will help you visualise the distribution of ages in the data set. The distribution can be over a variable, or it can also be over a period of time. Two of the most commonly used charts for visualising distribution are as follows:
* Histogram
* Scatter plots

Histograms are quite good at displaying the distribution of data over intervals, whereas scatter plots are good at visualising the distribution of data over two different variables.

Investigate the variation of prices of diamonds and note your observations.

<h3 style = "color : Brown"> Question </h3>

Perform a similar variation analysis on the quality of cut.

##### Solution

The observations in the variable cut are categories, not values like price or carat. So you will not be able to use a histogram to do a count analysis as you did earlier. For this kind of variable, you have to use a bar plot to compare the counts of observations in each category. You can use other plot types like pie chart doughnut chart, but it isn't easy to compare the counts of each class.

````python
# Before plotting extract the required data by performing necessary data manipulation
groupdata = data.groupby("cut").count()
groupdata
````

````python
# Create lists which hold the necessary data
x_vals = []
y_vals = []
for i in [0, 1, 4, 3, 2]:
    # The elements in the list are arranged in increasing order of the quality of cut.
    # This information can be gathered from the data dictionary.
    x_vals.append(groupdata.index[i])
    y_vals.append(groupdata.iloc[i,0])
````

````python
plt.bar(x_vals, y_vals)

# Adding and formatting title
plt.title("Frequency Plot: Cut of Diamonds\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Cut", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Count", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})

plt.show()
````

Observations:
1. The number of diamonds goes on increasing with increasing quality of cut.

<h3 style = "color : Brown"> Question </h3>

Represent the same information as above in a pie chart.

````python
explode = (0, 0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Very good')
plt.figure(figsize = [8,8])

plt.pie(y_vals, explode=explode, labels=x_vals, autopct='%1.1f%%',
        textprops={'fontsize': 16, 'fontweight' : 20, 'color' : 'Black'}, startangle=90)

# Adding and formatting title
plt.title("Distribution based on quality of cut\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})

plt.show()
````

As you can see, a pie chart, although visually attractive, is not a good tool to represent information. The changes in the different sectors are not noticeable. It is absolutely necessary to have written values. Now compare the same with the bar diagram earlier, the relative differences in the distribution are very clear even without the data labels.

The scatter plot with default settings looks very crowded, that is because there are a lot of data points, there are a couple of ways to solve this problem; you can either draw a smaller sample or use smaller and more transparent markers. You can also make the plot larger.

<h3 style = "color : Brown"> Question </h3>

Investigate the relationship of price with respect to the cut of the diamonds.

As you already know, "cut" is a categorical variable, and it will not be possible to use a scatter plot to analyse the variation. So you can explore the data in a lot of different ways like use colors to the display cut information in the above plot that is an excellent idea, you can try that on your own. You can also divide the data according to the cut and then plot box plots for each of the cuts. Such a plot will allow you to investigate the variation of statistical price data across various cuts.

````python
# Create the required dataframes
data_fair = data[data['cut'] == 'Fair']
data_good = data[data['cut'] == 'Good']
data_vgood = data[data['cut'] == 'Very Good']
data_premium = data[data['cut'] == 'Premium']
data_ideal = data[data['cut'] == 'Ideal']
````

````python
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']])
plt.show()
````

It looks like there are a lot of outliers in the data, which are making the data hard to analyse. Let's drop all the observations **above 12500 USD** as all of them are outliers.

````python
data_fair = data[(data['cut'] == 'Fair') & (data['price'] < 12500) ]
data_good = data[(data['cut'] == 'Good') & (data['price'] < 12500)]
data_vgood = data[(data['cut'] == 'Very Good') & (data['price'] < 12500)]
data_premium = data[(data['cut'] == 'Premium') & (data['price'] < 12500)]
data_ideal = data[(data['cut'] == 'Ideal') & (data['price'] < 12500)]
````

````python
plt.figure(figsize = [16,10])
plt.boxplot([data_fair['price'], data_good['price'], data_vgood['price'], data_premium['price'], data_ideal['price']])

plt.xticks((1,2,3,4,5), ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])

# Adding and formatting title
plt.title("Price vs Type of Cut\n", fontdict={'fontsize': 20, 'fontweight' : 20, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Type of Cut", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Price", fontdict={'fontsize': 12, 'fontweight' : 20, 'color' : 'Brown'})

plt.show()
````

Be cautious while making such modifications because the overall impact on the data is not very well explored yet, it might be possible that the average of different classes is affected differently. Luckily, in this case, the trend before and after the modification looks similar, but again caution is advised.

Observations:
1. The median price of the ideal cut is the least among all the five.
2. There are a lot of outliers in the ideal cut.

# **Summary**

Types of Charts and Graphs & Their Uses:

**Bar Chart:**


* Use: Excellent for comparing quantities across different categories.
* Good For: Visualizing differences in size between categories, such as sales figures across different regions or products.

**Scatter Plot:**

* Use: Ideal for examining the relationship or correlation between two continuous variables.
* Good For: Identifying trends, clusters, and outliers within paired numerical data.

**Line Graph:**

* Use: Best suited for displaying trends over time (time series data).
* Good For: Observing changes and trends, such as stock price movements over time or temperature changes throughout a year.

**Histogram:**

* Use: Used to depict the distribution of numerical data by showing the frequency of data points within certain range of values (bins).
* Good For: Understanding the distribution, spotting central tendencies, and identifying skewness in data sets.

**Box Plot:**

* Use: Provides a visual summary of several statistical parameters like the median, quartiles, and outliers.
* Good For: Comparing distributions and identifying outliers across different categories.

Additional Elements and Techniques Learned:
* Enhanced Plot Elements: Learned how to incorporate additional information into plots, such as labeling axes, titles, and using annotations to highlight specific data points or trends.

* Subplots: Mastered the technique of fitting several plots within a single figure, useful for comparing different datasets or variables effectively.

When to Use Specific Chart Types:

Comparison:

* Column and Bar Charts: When you need to highlight differences or make comparisons among discrete categories.

Composition:

* Pie/Doughnut Chart: Shows static composition of different parts to a whole at a specific time.
* Stacked Column Chart: Useful for displaying how parts of a whole change over time.

Relationship:

* Scatter Plot: Visualizes correlation between two variables and can help discover if they are related.
* Bubble Plot: Similar to scatter plots but adds a third dimension through the size of bubbles, indicating the magnitude or frequency of data.

Distribution:

* Histogram: Effective for showing the frequency of data points within certain ranges, helps in understanding the spread of numerical data.
* Scatter Plots for Distribution: Utilized to explore how data points are spread across two variables.
